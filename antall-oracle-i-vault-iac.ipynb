```python
def get_teamnavn(path):
    result =  re.search('/teams/(.*)/apps/', path)
    return result.group(1)
```

```python
def has_oracle(path):
    file = open(path).read()
    if 'oracle' in file:
        return 1
    return 0
```

```python
def has_pg(path):
    file = open(path).read()
    if 'postgresql' in file:
        return 1
    return 0
```

```python
df['team'] = df.apply(lambda x: get_teamnavn(x['path']), axis=1)
df['oracle'] = df.apply(lambda x: has_oracle(x['path']), axis=1)
df['pg'] = df.apply(lambda x: has_pg(x['path']), axis=1)
```

```python
teams = df.groupby('team').agg(antall_oracle=('oracle','sum'), antall_postgresql=('pg','sum')).reset_index()
nodb = teams[(teams['antall_oracle']==0) & (teams['antall_postgresql']==0)].reset_index(drop=True)
hasdb = teams[(teams['antall_oracle']>0) | (teams['antall_postgresql']>0)].reset_index(drop=True)
```
